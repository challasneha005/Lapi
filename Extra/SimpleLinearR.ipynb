# Simple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
#from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("area_price.csv")
print(df)
```

```python
%matplotlib inline
#import matplotlib.pyplot as plt
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(df.area,df.price,color="red",marker="*")
```

```python
l = linear_model.LinearRegression()
l.fit(df[['area']],df.price)
```

```python
l.predict([[3300]])
```

```python
print(l.coef_)
print(l.intercept_)
```

```python
#y=mx+b
print(l.coef_ * 3300 +l.intercept_)
```

```python
%matplotlib inline
#import matplotlib.pyplot as plt
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(df.area,df.price,color="red",marker="*")
plt.plot(df.area,l.predict(df[['area']]),color='blue')
```

```python
#for 5000
l.predict([[3300]])
print(l.coef_ * 5000 +l.intercept_)
```

**For new df**

```python
d = pd.DataFrame([1000,1500,2000,2500,3000,3500,4000],columns=['area'])
print(d)
```

```python
p = l.predict(d)
d['price'] = p
print(d)
```

```python
d.to_csv("predictsion.csv")
```
