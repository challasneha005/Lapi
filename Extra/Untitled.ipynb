```python
import string
import itertools
import csv

# Configuration
CHARS = string.ascii_uppercase + string.digits  # A-Z + 0-9
CODE_LENGTH = 14
BATCH_SIZE = 1000

def generate_batch_to_csv(batch_number):
    start_index = (batch_number - 1) * BATCH_SIZE
    end_index = batch_number * BATCH_SIZE

    all_codes = itertools.product(CHARS, repeat=CODE_LENGTH)
    batch_codes = itertools.islice(all_codes, start_index, end_index)

    filename = f"batch_{batch_number}.csv"

    with open(filename, mode="w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["S.No", "Code"])  # CSV header

        for i, code in enumerate(batch_codes, start=1):
            writer.writerow([i, ''.join(code)])

    print(f"Batch {batch_number} saved as {filename}")


# Example usage
generate_batch_to_csv(1)  # Batch 1 → first 1000 codes
generate_batch_to_csv(2)  # Batch 2 → next 1000 codes
```
