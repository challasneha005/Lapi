**DatFrame.apply**

```python
#pass a function and apply it to every single value of the Pandas series
#syn:DataFrame.apply(func, axis=0, broadcast=None, raw=False, reduce=None, result_type=None, args=(), **kwds) 
#axis : default 0, 0-index each col,1-col each row
#boardcast optional parameter
#raw:bool default false-passes each row or column as a Series to the function,
#if true-ndarray objects
import numpy as np
a=pd.DataFrame([[1,2]]*3,columns=['a','b'])
print(a)
print(a.apply(np.sqrt))
print(a.apply(np.sum,axis=0))
```

**DataFrame.aggregate**

```python
#aggregation to one or more column-sum,min,max
#syn:DataFrame.aggregate(func, axis=0, *args, **kwargs)
#**kwargs:keyword argument passed to function
#scalar=Series.agg single function
#Series=DataFrame.agg single function
#DataFrame=DataFrame.agg several functions
info=pd.DataFrame([[1,5,7],[10,12,15],[18,21,24],[np.nan,np.nan,np.nan]],columns=['X','Y','Z'])  
info.agg(['sum','min']) 
```

**DataFrame.cut**

```python
#sort the data values into bins
#convert a continuous variable to a categorical variable
#syn:pandas.cut(x, bins, right=True, labels=None, retbins=False, precision=3, include_lowest=False, duplicates='raise')  
a=pd.DataFrame({'num':np.random.randint(1,50,10)})
print(a)
a['bins']=pd.cut(x=a['num'],bins=[1,23,50])
print(a)
print(a['bins'].unique())
```

```python
a=pd.DataFrame({'num':np.random.randint(1,50,10)})
print(a)
a['bins']=pd.cut(x=a['num'],bins=[1,23,50],labels=['low','high'])
print(a)
print(a['bins'].unique())
```

**DataFrame.drop_duplicates**

```python
#common data cleaning task
#syn:DataFrame.drop_duplicates(subset=None, keep='first', inplace=False)  
#keep:consider duplicate values first:deletes other than 1st repeated value,last,False -deletes all
#inplace:boolean default False
emp = {"Name": ["Parker", "Smith", "William", "Parker"],"Age": [21, 32, 29, 21]} 
a=pd.DataFrame(emp)
print(a)
a=a.drop_duplicates()
print(a)
```
